Щоб змінити код, просто натисніть клітинку й почніть редагувати.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
import pandas as pd
from datetime import datetime, timedelta

# Завантаження даних із файла або бази даних
data = pd.read_csv('автопарк.csv')

# Інтервал ТО в кілометрах
то_interval_km = 10000

# Функція для розрахунку дати наступного ТО
def calculate_next_to_date(row):
    last_to_date = datetime.strptime(row['Дата_ТО'], '%Y-%m-%d')
    next_to_date = last_to_date + timedelta(days=то_interval_km / row['Пробіг_км'] * 365)
    return next_to_date

# Додавання стовпця з датою наступного ТО
data['Дата_наступного_ТО'] = data.apply(calculate_next_to_date, axis=1)

# Перевірка, чи наближається дата наступного ТО (наприклад, менше ніж через 30 днів)
today = datetime.now()
data['Наближається_ТО'] = data['Дата_наступного_ТО'] - today < timedelta(days=30)

# Виведення автомобілів, для яких наближається ТО
print(data[data['Наближається_ТО']])
```

```
import pandas as pd
from datetime import datetime, timedelta

# Завантажте дані з GPS трекерів (припустимо, що вони знаходяться у файлі "gps_data.csv")
# В цьому файлі повинні бути стовпці "Номер_автомобіля", "Пробіг_км" та "Дата_ТО"
df = pd.read_csv("автопарк.csv")

# Функція для обчислення наступної дати ТО на основі поточного пробігу та дати минулого ТО
def calculate_next_to_date(row):
    to_interval_km = 10000  # Встановіть необхідний інтервал пробігу для ТО
    last_to_date = datetime.strptime(row['Дата_ТО'], '%Y-%m-%d')
    next_to_date = last_to_date + timedelta(days=to_interval_km / row['Пробіг_км'] * 365)
    return next_to_date

# Додайте новий стовпець "Дата_наступного_ТО" і обчисліть його за допомогою функції
df['Дата_наступного_ТО'] = df.apply(calculate_next_to_date, axis=1)

# Виведіть результат
print(df)
```

```
import pandas as pd
from datetime import datetime, timedelta

# Завантажте дані з GPS трекерів (припустимо, що вони знаходяться у файлі "gps_data.csv")
# В цьому файлі повинні бути стовпці "Номер_автомобіля", "Пробіг_км" та "Дата_ТО"
df = pd.read_csv("автопарк.csv")

# Функція для обчислення наступної дати ТО та кілометражу, який має бути під час наступного ТО
def calculate_next_to_data(row):
    to_interval_km = 10000  # Встановіть необхідний інтервал пробігу для ТО
    last_to_date = datetime.strptime(row['Дата_ТО'], '%Y-%m-%d')
    next_to_date = last_to_date + timedelta(days=to_interval_km / row['Пробіг_км'] * 365)
    next_to_km = row['Пробіг_км'] + to_interval_km
    return pd.Series({'Дата_наступного_ТО': next_to_date, 'Кілометраж_наступного_ТО': next_to_km})

# Використовуйте apply, щоб виконати функцію для кожного рядка
df[['Дата_наступного_ТО', 'Кілометраж_наступного_ТО']] = df.apply(calculate_next_to_data, axis=1)

# Виведіть результат
print(df)

# Тепер у вас є DataFrame з інформацією про дату та кілометраж наступного ТО для кожного автомобіля
# Ви можете використовувати цю інформацію для надсилання нагадувань чи подальших дій.
```
